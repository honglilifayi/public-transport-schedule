```python
import codecs
import pandas as pd
import numpy as np
import csv
from StringIO import StringIO
from datetime import datetime
```

```python
with codecs.open('stoptimes.csv', 'r') as f:
    data = f.read()
```

```python
stoptimes = pd.read_csv(StringIO(data.replace('"', '')), sep=',', encoding='utf-8', quoting=csv.QUOTE_NONE)
```

```python
with codecs.open('stops.csv', 'r') as f:
    data2 = f.read()
stops = pd.read_csv(StringIO(data2.replace('"', '')), sep=',', encoding='utf-8', quoting=csv.QUOTE_NONE)
stops.dtypes
stops.head()
```

```python
type(stoptimes)
```

```python
# Merge https://chrisalbon.com/python/data_wrangling/pandas_join_merge_dataframe/
from IPython.display import display
from IPython.display import Image
```

```python
merged = pd.merge(stoptimes, stops, on='stop_id', how='inner')
merged.head()
```

```python
# dropped columns https://stackoverflow.com/questions/13411544/delete-column-from-pandas-dataframe-using-del-df-column-name
new = merged.drop(merged.columns[[5, 6, 7,8]], axis=1)
new.head()
```

```python
# reorder columns https://stackoverflow.com/questions/41968732/set-order-of-columns-in-pandas-dataframe/41968825
new = new[['stop_id', 'stop_name', 'arrival_time', 'departure_time', 'trip_id', 'stop_sequence', 'stop_lat', 'stop_lon']]
new
```

```python
new.info()
```

```python
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.drop_duplicates.html
stoptime = new.drop_duplicates(subset='stop_id', keep='first', inplace=False)
stoptime.shape
```
